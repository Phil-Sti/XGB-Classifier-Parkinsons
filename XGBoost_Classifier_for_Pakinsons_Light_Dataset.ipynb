```python
# Read the data using Pandas to display the columns. First, we use the describe function to show usefull statistics and information about the dataset.
df = pd.read_csv('/content/parkinsons.data', sep=",")
df.describe()
```

```python
# Let's get the features and labels: the status column will act as label: 1 is parkinson positive, 0 is negative.
features = df.loc[:, df.columns != 'status'].values[:, 1:]
labels = df.loc[:, 'status'].values
```

```python
# Scale the features between -1 and 1:
scaler = MinMaxScaler((-1, 1))
x = scaler.fit_transform(features)
y = labels

# Split the dataset between Training and Test sets. The test set will be 20% of the overall dataset. Also, we will shuffle the row to avoid bias:
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=7)
```

```python
# Train the model using XGB Classifier
model = XGBClassifier()

# define the eval set and metric
eval_set = [(x_train, y_train), (x_test, y_test)]
eval_metric = ["auc","error"]

# We fit the model (verbosity can be set to False to avoid the full epochs' log)
model.fit(x_train, y_train, eval_metric=eval_metric, eval_set=eval_set, verbose=True)
```

```python
# Last, we calculate the model's accuracy using the .predict function:
test_pred = model.predict(x_test)
train_pred = model.predict(x_train)
print('Train Accuracy: ', accuracy_score(y_train, train_pred))
print('Test Accuraccy: ', accuracy_score(y_test, test_pred))
print("\n")
print(f"The model's actual accuracy is {accuracy_score(y_test, test_pred) * 100} %")
```
